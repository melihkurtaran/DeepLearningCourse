## Functions (def...return)

````python
# Define the function 'by2' that multiplies by two

# If the value of x is not passed to the call, then it will be its default value
# (10 in this case). Setting a default value is optional.
def by2(x=10):
    return 2 * x

x = 4
y = by2(x)
print(f"Twice {x} is {y}")
````

````python
# If the function can take a variable number of arguments, then *args is set.
# 'args' will be a tuple of arguments. The truthiness of a tuple is True if it
# contains at least one element, so you just have to check args.

def my_function_1(x=10, *args):
    output = x
    if args:
        for arg in args:
            output += arg
    return output


x = 1; y = 2; z = 3

# This is the function call with two additional arguments passed by position
salida = my_function_1(x, y, z)
print(f"%i + %i + %i = %i" %(x, y, z, salida))
````

````python
# If the function also has to be able to receive arguments passed by name, then
# **kwargs is added. 'kwargs' is a dictionary of the parameters passed by name

def my_function_2(x=10, *args, **kwargs):
    output = x

    # Get value of the key "oper" in the dictionary kwargs if it exists, or set
    # default otherwise
    op = kwargs["oper"] if "oper" in kwargs else "multiply"

    # You can also do it with setdefault, but in this case the value is included
    # in the dictionary
    # op = kwargs.setdefault("oper", "multiply")

    print(kwargs)
    # Use all values in args to calculate output
    if args:
        for arg in args:
            if op == "multiply":
                output *= arg
            elif op == "divide":
                output /= arg
            elif op == "subtract":
                output -= arg
            else:
                output += arg

    return output

print("Result from multiply operation: {}".format(my_function_2(10, 9, 8, 7)))

print("Result from divide operation: {}".format(
    my_function_2(10, 9, 8, 7, oper="divide")))

print("Result from subtract operation: {}".format(my_function_2(
    10, 9, 8, 7, oper="subtract")))

print("Result from subtract operation: {}".format(my_function_2(
    10, 9, 8, 7, oper="sum")))
````

## Classes/objects

In python everything is an object, with its attributes (properties and methods). Unlike other languages, direct assignments can be made. All classes have a `__init__()` function whose first parameter is always _'self'_ (it is a reference to the current instance, i.e. the object being initialized).

The `__init__()` function initializes the values of the current instance of the current class.

````python
# Class definition
class Vehicle:
    # __init__ function
    def __init__(self, license_plate, mileage, color):
        self.license_plate = license_plate
        self.mileage = mileage
        self.color = color
        print("__init__ function from Vehicle executed!")

    # Other methods
    def route(self, miles):
        self.mileage += miles

    def show_data(self):
        v = type(self).__name__.lower()
        print("The {} with license plate {} is {} and has {} miles".format(
            v, self.license_plate, self.color, self.mileage))

# Instantiate an object of this class
v1 = Vehicle("7777-HMV", 0, "black")
# Show the data
v1.show_data()

# Change the color
v1.color = "red"
# Make a route of 40 miles
v1.route(40)
# And show the data again
v1.show_data()
````

````python
# It is very common for classes to be created from other already established
# classes. The way to do it is very simple, you just have to pass the parent
# class as a parameter when defining the child class. In this way all the
# methods and properties of the parent class are imported (that is, the
# __init__ function of the parent class is executed)

class Car(Vehicle):
    pass

v2 = Car("8888-HMV", 1200, "green")
v2.route(1000)
v2.show_data()
````

````python
# If a new __init__ function is defined in the class, then the one in the
# parent class is not executed. In this cases it might be interesting to import
# the methods and properties of the parent class. This can be done directly
# executing the __init__ function of the parent class
# "ParentClassName.__init__()" or substituting this name by super()

class Truck(Vehicle):
    def __init__(self, license_plate, mileage, color, height):
        super().__init__(license_plate, mileage, color)
        self.height = height
        print("__init__ function from Truck executed!")

    # And in this case we override the show_data method because we want to add
    # some functionality
    def show_data(self):
        v = type(self).__name__.lower()
        print("The {} with license plate {} is {}, has {} miles".format(
            v, self.license_plate, self.color, self.mileage) +
            ", and it is {} feet tall.".format(self.height))

c1 = Truck("9999-HMV", 25000, "white", 13.5)
c1.show_data()
````

Collections in python are iterable objects in the sense that they allow you to loop through all of their elements.

Iterables in python allow you to create iterators using the iter() function. Iterators are a subclass of the iterable class, and are objects that return the next element in the collection each time they are called with the next() function.

````python
# A list is an iterable object, but it is not an iterator or a generator.
list_1 = [0, 1, 2, 3, 4]

# An iterable object has a private method called __iter__() that can be called
# through the function iter(). Check that lists has the __iter__() method.
print(list(filter(lambda x: x.startswith("__i"), dir(list_1))), "\n")
````

````python
# An iterator is a subclass of an iterable object. The way to create it is by
# calling the iter() function. The iterator incorporates the __next__() function
# that allows access to the next element of the collection, and can be called
# through the next() function
iterator_1 = iter(list_1)

# Check if __next__() function exists
print(list(filter(lambda x: x.startswith("__n"), dir(iterator_1))), "\n")

# Check the class of the object iterator_1
print(type(iterator_1))
# All collections in python are iterable, so they incorporate the iter() function that allows you to create iterators
````

````python
# When we do a for loop, internally an iterator is created and each element is
# called through the __next()__ function
for i, n in enumerate(list_1):
    print(f"Element {i}: {n}")

# We can emulate this behavior with the next() function applied on the iterator
print("\nElement 0:", next(iterator_1))
print("Element 1:", next(iterator_1))
````

````python
# If the end is reached and an attempt is made to access the next element, an
# exception occurs
print("Element 2:", next(iterator_1))
print("Element 3:", next(iterator_1))
print("Element 4:", next(iterator_1))

# try-except is used to catch errors. You'll see in detail later. By now, it's
# enough to know that it'll execute the except block only if there's an error in
# the try block.
try:
    print("Element 5:", next(iterator_1))
except:
    print("Exception: No more elements to loop over")
````

````python
# The 'for' and 'while' functions can work interchangeably with iterables and
# iterators

# list_1 is a list, so it's an iterable object.
for n in list_1:
    print(n)

print()

# iterator_1 is an iterator object.
iterator_1 = iter(list_1)
for n in iterator_1:
    print(n)
````

````python
# A custom iterator can be easily created by creating a class with the
# __iter__() and __next__() functions.

class MyIterator:
    def __iter__(self):
        self.value = 1
        return self  # Iterators created like this return themselves

    def __next__(self):
      if self.value <= 4:
          output = self.value
          self.value += 1
          return output
      else:
          raise StopIteration

# An object of class MyIterator is created
obj1 = MyIterator()

# Iterates in a for loop
for x in obj1:
  print(x)

print()
# Or create the iterator of MyIterator class with iter() and iterate with next()
obj2 = iter(MyIterator())
print(next(obj2))
print(next(obj2))
````

````python
# A generator is a subclass of an iterator, and behaves reasonably similar to
# it, although the generator is more memory efficient.

# To create a generator just create a function that returns values with 'yield'
# instead of return. Each time the function is called, it returns a value with
# 'yield' and saves the state of the function at that point, and returns to it
# on the next call.

def generator_num():
    i = 0
    k = 1
    yield i  # First value to return. Do not go through this point again.
    i += 1
    yield i  # Second value to return. Do not go through this point again.
    while i > 0:
        if i == 4: k = -1
        i += k
        yield i  # Rest of values returned in each iteration of the 'while' loop

generator = generator_num()
print(type(generator), "\n")

for i in generator:
    print(i)
````

````python
# Comparison between the memory usage of an iterator and a generator in a
# simple list

# Try different values of the variable maxim
maxim = 100

# Do not modify this cell below this line
# ________________________________________________

list1 = list(range(maxim))

iterator1 = iter(list1)

# The iterator doesn't take up much, but the list will remain in memory until
# the iterator is destroyed, so count it in memory usage.
print("Memory used by the iterator: {}".format(
    iterator1.__sizeof__() + list1.__sizeof__()))

# Generator of the same list
def generator1(maxim=10):
    i = 0
    yield i
    while i < maxim:
        i += 1
        yield i

gen = generator1(maxim)

# In the case of the generator, it is only necessary to compute what it uses.
print(f"Memory used by the generator: {gen.__sizeof__()}")
````

## Scope of variables
Variables are available within the region in which they are created, and in all subregions of it. That is, if a variable is created inside a function, then it is available in that function and in all the functions that are defined inside this function.

````python
try:
    del x  # Makes sure that the variable x does not exist
except:
    pass
````

````python
# Variable x defined and used inside a function
def func():
    x = 1
    print(x)  # x exists inside this function

func()
````

````python
try:
    print(x)  # x doesn't exist outside the function
except:
    print("This variable doesn't exist")
````

````python
# Variable x defined inside a function and used in a subfunction
def func_1():
    x = 1
    def func_2():
        print(x)
    func_2()

func_1()  # x exists inside func_1 and inside func_2
````

````python
# If the variable is defined in the main body of the program, then it is global
# even if it is not explicitly stated.

x = 2  # Global variable

def func_1():
    print(x)
    
func_1()
````

````python
# If the variable is redefined with the same name inside the function, then
# python considers that there are two variables, one with global scope and one
# with local scope.

x = 2  # Global variable

def func_1():
    x = 1
    print(x)
    
func_1()  # It prints the local variable
print(x)  # It prints the global variable
````

````python
# If you want to modify the global variable from within a function, you have to
# explicitly indicate it with global

x = 2  # Global variable 

def func_1():
    global x
    x = 1
    print(x)
    
func_1()  # It prints the local variable (that is also the global one)
print(x)  # It prints the global variable
````

````python
# If there are nested functions, any local definitions locally override
# definitions of higher scopes

x = 3  # Global variable
def func_1():
    # Local variable within func_1 and available within any subfunction of it,
    # if not redefined
    x = 2
    def func_2():
        # Local variable definition inside func_2. Within this function, the
        # definition of x from func_1 is no longer valid, since there's a new
        # definition
        x = 1
        print("x of func_2:", x)
        return x
    _ = func_2()
    print("x of func_1:", x)
    
func_1()
````

````python
# If you want to assign values to x of func_1 from inside func_2, you can use
# the keyword 'nonlocal' to indicate it

x = 3  # Global variable
def func_1():
    # Local variable within func_1 and available within any subfunction of it,
    # if not redefined
    x = 2
    def func_2():
        # Non local redefinition of variable x from func_1
        nonlocal x
        x = 1
        print("x of func_2:", x)
        return x
    _ = func_2()
    print("x of func_1:", x)
    
func_1()
````

Remember that you can know all the global and local variables with the built in functions:

* globals()
* locals()

And you can use the keywords `global` and `nonlocal` to specify the scope.

## Modules

A module is a library of code. To create a module just save the generated code in a .py file

For example, save the following code in a file called myModule.py:

```python
def greeting(name):
    print("Hello {}".format(name))

def goodbye(name):
    print("Bye {}".format(name))
```

Then in your code you can import this whole module by simply putting

```python
import myModule
```

and if you want to use a function from that module you will have to use the syntax

```python
myModule.greeting("Mary")
```

Similarly, if the module contains variables they can be imported. Thus, if the module is the following:

```python
person_1 = {name = "Mary",
            surname = "Doe",
            age = 29}

def greeting(name):
    print("Hello {}".format(name))

def goodbye(name):
    print("Bye {}".format(name))
```

You could import the module and use the variable like this:

```python
import myModule

my_name = myModule.person_1["name"]
myModule.greeting(my_name)
```

However, it might be that the module name is not practical. What can be done is to use an alias at import time:

```python
import myModule as mM  # mM is a synonym for myModule in this code

my_name = mM.person_1["name"]
mM.greeting(my_name)
```

Finally, sometimes it may be preferable to import only one single function or variable from the module. This can be done in the following way:

```python
from myModule import greeting
```

Note: To list all the functions of a module you can use the `print(dir(myModule))` command after importing it.

## try...except

It is used for error handling. The `try` block is for executing test code. If an error occurs, the `except` code for the error in question is executed (if the type of error is given). If no error occurs, the code in the `else` block is executed. The `finally` block is always executed. The `else` and `finally` blocks are optional.

````python
try:
    print(x)
except NameError:  # Predefined exception
    print("Variable x not defined. Default value is set.")
    x = "Mary"
except:
    print("Other exception ocurred!")
else:
    print("If no error has occurred, you get here")
finally:
    print("This code is executed regardless of whether errors occur or not")
    print(x)
````

The programmer can define his own error codes and raise the corresponding exception with the `raise` statement.

````python
x = "Hello"

if not type(x) is int:
    raise TypeError("An integer number is required")
````

## Decorators

This explanation is partially based on the one that you can find in the next link. See the original one, that is longer and contains more examples, at: https://realpython.com/primer-on-python-decorators/#syntactic-sugar

Check a good example of decorator for logging at: https://book.pythontips.com/en/latest/decorators.html?highlight=decorators#logging

In python everything is an object. Functions and methods are objects too. Since they can be called, they are called _callable_. In fact, any object that implements the `__call__()` method can be called a callable.

If you have a function `fcn` you can call it with the sintaxis `fcn()`. If you use it without the parentheses, then you are passing a reference to that function.

```python
result_of_function_call = fcn()
reference_to_function = fcn
```

````python
# Define a simple function that adds 1 to any number
def add1(input):
    return 1 + input
````

````python
# Call this function with input parameter equals to 2, so it should return 3
result = add1(2)  # function call
print(result)
````

````python
# Now we assign to another variable a reference to that function
also_add1 = add1  # function reference

# As we can see, this new variable is a function. In fact, also_add1 is just
# another identifier (function name) that points to the same function
print(also_add1)
````

````python
# So much so, that we can call it interchangeably with either of both names.
result = add1(2)
print(result)
result = also_add1(2)
print(result)
````

Decorators are a way for python to be able to add some functionality to a  function without having to modify it. Take a look at the function below, which we are now going to discuss in detail.

````python
# A 'decorator_fcn' function is defined that receives the 'fcn' function as a
# parameter.

# This decorator_fcn is the decorator function, and its parameter 'fcn' is the
# function to be decorated.
def decorator_fcn(fcn):
    # Inside 'decorator_fcn', another local function 'inner_fcn' is defined
    def inner_fcn():
        print("Something to do before the function")
        # A call to the original function received as parameter
        fcn()
        print("Something to do after the function")
    # The 'decorator_fcn' function returns a reference to the 'inner_fcn'
    # function
    return inner_fcn
````

Although it is commented in the code, the following are the relevant aspects to take into account about the previous code:

* The function 'decorator_fcn' receives as parameter 'fcn' a reference to a function.

* The 'decorator_fcn' function defines a local function inside it, called 'inner_fcn'. This implies, that the 'inner_fcn' function will be defined **only** when 'decorator_fcn' is called.

* The function 'fcn' is used as a function inside 'inner_fcn'

* 'inner_fcn' does something, but currently it returns nothing.

* 'decorator_fcn' returns a reference to 'inner_fcn'

Now we define another function. This is a first-order function that does a simple thing: it just prints some text.

````python
def say_hello():
    print("Helloooo")
````

````python
# Call the function
say_hello()
````

Pay attention to the following code, which is directly a decoration, although later we will see another syntax to do the same.

The behavior of the 'say_hello' function is being modified. The original 'say_hello' function itself is now passed as a parameter to the decorator function.

The result, as you might guess, is that from now on 'say_hello' will do what 'inner_fcn' does, which means, doing what the original 'say_hello' function did plus some extra functionality before and after.

````python
# Decoration of say_hello function
say_hello = decorator_fcn(say_hello)
````

````python
# Once we've decorated say hello, we can call it to check that we have the new
# functionality
say_hello()
````

````python
# Well, what you have done now can be done with another syntax and it implies
# the same. Specifically, at the time of defining the 'say_hello' function
# (which is the decorated function) you can indicate which function should be
# the decorator function. It is done like this:

@decorator_fcn
def say_hello():
    print("HellooooOOOooo")
````

````python
# See the new functionality when you call say_hello() once decorated
say_hello()
````

Now we define a new function that we are going to decorate; but with the caveat that this new function takes a couple of parameters and returns a result.

We show it as text because we still haven't defined the decorator function.

```python
# Define a function that returns the mean of two numbers
def mean_of_2(n1, n2):
    return (n1 + n2) / 2
```

````python
# The decorator function will inform of the values of the input parameters
# before calculating the mean, and also will calculate the square of the
# output

def dec_fcn(fcn):
    # See that in this case the inner_fcn can get input parameters
    def inner_fcn(*args):
        # Functionality before the fcn call
        print("Number 1: {} - Number 2: {}".format(args[0], args[1]))
        
        # fcn call and get the mean
        mn = fcn(*args)
        
        # Functionality after the fcn call
        # Use the mean to calculate the square mean
        sq_mn = mn**2

        # Return the result of the fcn and the extra functionality
        return mn, sq_mn
    
    return inner_fcn
````

````python
# And now, let's define the function with its decorator
@dec_fcn
def mean_of_2(n1, n2):
    return (n1 + n2) / 2
````

````python
# And check if it works as desired
mn, sq_mn = mean_of_2(2, 4)

print("Mean: {} - Squared mean: {}".format(mn, sq_mn))
````

````python
# One last detail may be important depending on what you do. The function is
# confused about who it is. It believes to be 'inner_fcn'.
mean_of_2.__name__
````

````python
import functools

# This could be a problem in some circumstances; to fix it, just use
# the @functools.wraps() decorator before defining the inner function, which
# updates a few introspection attributes (__name__, __doc__).

def dec_fcn(fcn):
    # Decorator to update introspection attributes
    @functools.wraps(fcn)
    def inner_fcn(*args):
        # Functionality before the fcn call
        print("Number 1: {} - Number 2: {}".format(args[0], args[1]))
        
        # fcn call and get the mean
        mn = fcn(*args)
        
        # Functionality after the fcn call
        # Use the mean to calculate the square mean
        sq_mn = mn**2

        # Return the result of the fcn and the extra functionality
        return mn, sq_mn
    
    return inner_fcn
````

````python
# Delete the function
del mean_of_2

# Re-define it
@dec_fcn
def mean_of_2(n1, n2):
    return (n1 + n2) / 2

# And check if it works as desired
mn, sq_mn = mean_of_2(2, 4)

print("Mean: {} - Squared mean: {}".format(mn, sq_mn))

# And check its name
print("Name: {}".format(mean_of_2.__name__))
````
